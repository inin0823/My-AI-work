```python
#import request 套件抓取PPT電影版的網頁原始碼(HTML)
import urllib.request as req
url="https://www.ptt.cc/bbs/movie/index.html"

#建立一個Request物件，附加Resquest Headers的資訊
request=req.Request(url, headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:73.0) Gecko/20100101 Firefox/73.0"
}) #讓我們看起來像一個普通的使用者，最重要的是"User-Agent"
with req.urlopen(request) as response:
    data=response.read().decode("utf-8")
print(data)
```

```python
#解析原始碼，取得每篇文章的標題
import bs4
root=bs4.BeautifulSoup(data,"html.parser")
print(root.title.string) #抓標籤裡面的文字
```

```python
#尋找class=title"的div標籤
titles=root.find("div",class_="title") #"div"表示想要找的標籤名稱 ，class_="title"表示篩選的條件
print(titles.a.string) #抓到其中一個內容
```

```python
#尋找"所有"class=title"的div標籤
titles=root.find_all("div",class_="title") #"div"表示想要找的標籤名稱 ，class_="title"表示篩選的條件
for title in titles:
    if title.a !=None:
        print(title.a.string)
```
